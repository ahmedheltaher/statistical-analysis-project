```python
import numpy as np
import pandas as pd
import plotly.express as px
from matplotlib import pyplot as plt
from collections import Counter
from scipy import stats
```

```python
plt.style.use('bmh')
plt.rcParams["figure.figsize"] = (15,10)

heart_data = pd.read_csv("https://raw.githubusercontent.com/ahmedheltaher/statistical-analysis-project/master/heart.csv")
```

```python
counts = Counter(heart_data.FastingBS[:-1])

plt.pie(counts.values(), labels = ["Fasting Blood Pressure Normal", "Fasting Blood\nPressure > 120 mg/dl"], autopct = '%1.1f%%',)
plt.axis('equal')
plt.title('Fasting State')
plt.show()
```

```python
fig = px.histogram(heart_data["FastingBS"], template = "seaborn")
fig.show()
```

```python
heart_data['Cholesterol'].hist() 
plt.title('Cholesterol')
plt.xlabel('Cholesterol')
plt.ylabel('Number of people')
plt.show()
```

```python
heart_data['Age'].hist() 
plt.title('Age Frequency')
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.show()
```

```python
fig = px.histogram(heart_data, x='Age', marginal='box', template = "seaborn")
fig.show()
```

```python
age_data = heart_data["Age"].to_numpy()
age_mean = np.mean(age_data)
age_median = np.median(age_data)
age_mode = stats.mode(age_data)
age_standart_deviation = np.std(age_data)
print(age_mean,"--",age_median,"--",age_data[54],'\n',age_standart_deviation,"--",stats.skew(age_data))
```

```python
sentiment = []
for row in heart_data["Sex"]:
    sentiment.append(row[0].split()[0])

counts = Counter(sentiment[:-1])
plt.pie(counts.values(), labels = ["Male", "Female"], autopct = '%1.1f%%',)
plt.axis('equal')
plt.title('Gender Frequency')
plt.show()
```

```python
fig = px.bar(heart_data["Sex"] , template = "seaborn")
fig.show()
```
